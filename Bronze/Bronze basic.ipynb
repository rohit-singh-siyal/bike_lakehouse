Read from csv file


```python
df = ( spark.read.format("csv")
      .option("path", "/Volumes/workspace/bronze/source_systems/source_crm/cust_info.csv")
      .option("header", True)
      .option("inferschema", True)
      .load()
)
display(df)
```

Write it to Bronze layer

```python
df.write.mode("overwrite").saveAsTable("workspace.bronze.crm_cust_info")
```
